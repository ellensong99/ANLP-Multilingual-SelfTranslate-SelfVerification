```python
import openai
import pandas as pd

# Load your data
data = pd.read_excel(questions_file_path_150_zh)
data.columns = ['Math Problem', 'Ground Answer','Calculated Answer_a','Calculated Answer_b']

# Function to send each math problem to the OpenAI API
def get_answer(problem):
    #formatted_problem = f"{problem} Please answer the question in English and then write the final numerical answer after the format 'The answer is:'."
    formatted_problem = f"{problem} 请以'答案是：'的格式写出最终的数字答案。"
    chat_completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "user",
                "content": formatted_problem
            }
        ],
    )
    print(chat_completion)
    # Assuming the last message from the model is the answer
    text = chat_completion.choices[0].message.content
    # Extracting only the numeric answer
    #match = re.search(r"The answer is: [\$]?([\d,]+(?:\.\d+)?)", text, re.IGNORECASE)
    match = re.search(r"(?:The answer is:|答案是：) ?[\$]?([\d,]+(?:\.\d+)?)", text, re.IGNORECASE)
    #match = re.search(r"答案是：(\d+)", text)
    if match:
        return str(match.group(1))
    return None

# print(get_answer("杰罗姆有 4 个朋友，他们在某一天来拜访他。在杰罗姆开门前，第一位朋友按了 20 次门铃，第二位朋友比杰罗姆的第一位朋友多按了 1/4 次门铃。第三位朋友比第四位朋友多按了 10 次门铃。如果第四位朋友按了 60 次门铃，门铃响了几次？"))

# Apply the function to the 'Math Problem' column
data['Calculated Answer_a_clean'] = data['Math Problem'].apply(get_answer)

# Save the results back to an Excel file
data.to_excel(questions_file_path_150_zh, index=False)
```

```python
import openai
import pandas as pd

# Load your data
data = pd.read_excel('/content/drive/My Drive/ANLP_HW4/questions_only.xlsx')
data.columns = ['question', 'answer']

# Function to send each math problem to the OpenAI API
def get_answer(problem):
    formatted_problem = f"{problem} Please write the final numerical answer after the format 'The answer is:'."
    chat_completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {
                "role": "user",
                "content": formatted_problem
            }
        ],
    )
    print(chat_completion)
    # Assuming the last message from the model is the answer
    text = chat_completion.choices[0].message.content
    # Extracting only the numeric answer
    match = re.search(r"The answer is: [\$]?([\d,]+(?:\.\d+)?)", text, re.IGNORECASE)
    if match:
        return str(match.group(1))
    return None

# Apply the function to the 'Math Problem' column
data['Calculated Answer_c'] = data['question'].apply(get_answer)

# Save the results back to an Excel file
data.to_excel('/content/drive/My Drive/ANLP_HW4/questions_only.xlsx', index=False)
```
